# Exploração interativa do pipeline

Este notebook mostra como usar o script `analise_histopatologia.py` como apoio didático.

**Aviso:** o pipeline é exploratório e não diagnóstico.

```python
from analise_histopatologia import load_image, normalize_image, extract_nuclear_channel, segment_nuclei, compute_metrics, overlay_nuclei, create_label_visualization
import matplotlib.pyplot as plt

image_path = '../examples/sua_imagem.png'  # ajuste aqui
img = load_image(image_path)
img_norm = normalize_image(img)
mask = extract_nuclear_channel(img_norm)
labels = segment_nuclei(mask)
metrics = compute_metrics(img_norm, labels)
overlay = overlay_nuclei(img_norm, labels)
label_vis = create_label_visualization(labels)
metrics
```

```python
fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img_norm)
ax1.set_title('Imagem normalizada')
ax1.axis('off')

ax2 = fig.add_subplot(2,2,2)
ax2.imshow(mask, cmap='gray')
ax2.set_title('Máscara nuclear')
ax2.axis('off')

ax3 = fig.add_subplot(2,2,3)
ax3.imshow(label_vis)
ax3.set_title('Segmentação colorida')
ax3.axis('off')

ax4 = fig.add_subplot(2,2,4)
ax4.imshow(overlay)
ax4.set_title('Overlay')
ax4.axis('off')

plt.tight_layout()
plt.show()
```
